```python
localidades = []

for feature in pt_geo['features']:
        localidades.append(feature)
```

```python
# In order to feed the GeoJson into Plotly we need to make a small alteration,
# In this case add the 'id' key to each feature containing the value of each countries ISO-A3 code 
# Which is stored in the key 'feature['properties']['ADMIN']'
id_list = []

for feature in localidades:
    feature['id'] = feature['properties']['name_2']   # or 'ISO-A3'
    id_list.append(feature['properties']['name_2'])
```

```python
pt_geo = dict(type='FeatureCollection', features=localidades)
```

```python
id_list
```

```python
data_choropleth_pt = dict(type='choroplethmapbox',
                              geojson=pt_geo,
                              locations=coordenadas_covid['concelho'],
                              z=np.log(coordenadas_covid['incidencia']),
                              colorscale='inferno',
                              colorbar=dict(title='covid 19'))

layout_choropleth_pt=dict(mapbox=dict(style='white-bg',
                                          layers=[dict(source=feature,
                                                       below='traces',
                                                       type='fill',
                                                       fill=dict(outlinecolor='gray'))
                                                  for feature in pt_geo['features']],
                                          center=dict(lat=coordenadas_covid['lat'].iloc[0],
                                                      lon=coordenadas_covid['lng'].iloc[0]),
                                          zoom=5.5),
                              title=dict(text='Covid Incidence in Portugal',
                                         x=.5))

fig_choropleth_pt=go.Figure(data=data_choropleth_pt,layout=layout_choropleth_pt)
fig_choropleth_pt.show(renderer='browser')
```
